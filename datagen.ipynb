```python
import pandas as pd
import numpy as np

# Simulate 20 heat wave entries
np.random.seed(42)

heat_wave_data = {
    "Disaster Subtype": ["heat wave"] * 20,
    "temp": np.round(np.random.uniform(35, 42, 20), 1),
    "humidity": np.round(np.random.uniform(10, 30, 20), 1),
    "precipitation": [0.00] * 20,
    "windgust": np.round(np.random.uniform(15, 30, 20), 1),
    "windspeed": np.round(np.random.uniform(10, 25, 20), 1),
    "pressure": np.round(np.random.uniform(1008, 1014, 20), 1)
}

heatwave_df = pd.DataFrame(heat_wave_data)
heatwave_df
```

```python
df_original = pd.read_csv("cleaned_data_with_weather_data .csv")
df_augmented = pd.concat([df_original, heatwave_df], ignore_index=True)
df_augmented.to_csv("disaster_weather_augmented.csv", index=False)
df_augmented
```
